### Informações sobre os campos do dataset

As informações dos campos foram coletadas no próprio Kaggle:

- `0 - Country` : Nome do país
- `1 - Year` : Ano da observação
- `2 - Status` : Desenvolvimento ou status de desenvolvimento do país (Desenvolvido ou em Desenvolvimento)
- `3 - Life expectancy` : Expectativa de vida em anos
- `4 - Adult Mortality` : Taxas de mortalidade adulta para ambos os sexos (probabilidade de morrer entre 15 e 60 anos por 1000 habitantes)
- `5 - infant deaths` : Número de mortes de bebês por 1000 habitantes.
- `6 - Alcohol` : Consumo per capita de álcool registrado (15 anos ou mais) em litros de álcool puro. - Percentual de despesa: Despesa com saúde como porcentagem do Produto Interno Bruto per capita (%).
- `7 - percentage expenditure` : Despesa com saúde como porcentagem do Produto Interno Bruto per capita (%).
- `8 - Hepatitis B` : Cobertura de imunização contra a Hepatite B (HepB) entre crianças de 1 ano (%).
- `9 - Measles` : Número de casos de sarampo notificados por 1000 habitantes.
- `10 - BMI` : IMC médio da população.
- `11 - under-five deaths` : Número de mortes de crianças menores de cinco anos por 1000 habitantes.
- `12 - Polio` : Cobertura de imunização contra a Pólio (Pol3) entre crianças de 1 ano (%).
- `13 - Total expenditure` : Despesa do governo geral com saúde como porcentagem da despesa governamental total (%).
- `14 - Diphtheria` : Cobertura de imunização contra Difteria, Tétano e Coqueluche (DTP3) entre crianças de 1 ano (%).
- `15 - HIV/AIDS` : Mortes por HIV/AIDS por 1000 nascidos vivos (0-4 anos).
- `16 - GDP` : Produto Interno Bruto per capita (em USD).
- `17 - Population` : População do país.
- `18 - thinness 1-19 years` : Prevalência de magreza entre crianças e adolescentes de 10 a 19 anos (%).
- `19 - thinness 5-9 years` : Prevalência de magreza entre crianças de 5 a 9 anos (%).
- `20 - Income composition of resources` : Índice de Desenvolvimento Humano em termos de composição de renda dos recursos (índice variando de 0 a 1).
- `21 - Schooling` : Número de anos de escolaridade (em anos).

### Realizando análises iniciais

```
# Ajuste das variávies com espaço no rótulo (" ")
life_expectancy_data.columns = life_expectancy_data.columns.str.strip()
```

#### Evoluções ao longo do anos

```
variavel_evolucao = 'Life expectancy'
ano_a_ano = life_expectancy_data.groupby('Year')[variavel_evolucao].mean()

plt.figure(figsize=(12, 6))
plt.plot(ano_a_ano, marker='o', color='#FF9999', linestyle='-', linewidth=2)
plt.title('Evolução média ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel(variavel_evolucao)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

É visível que a expectativa de vida aumentou ao longo dos anos. Caso alteremos a variável de evolução no código acima para outras (ex: trocar "Life expectancy" para "Polio"), percebe-se que fatores de imunização apresentam evolução justificando o aumento. Escolaridade e fatores sociais também apresentam evolução podendo estar relacionados.

#### Status de países

```
# Distribuição por status
cores = ['#FF9999', '#66B2FF']

life_expectancy_data['Status'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors = cores)
plt.title(f'Distribuição de Status')

plt.show()
```

Neste dataset há uma quantidade consideravelmente maior de países que ainda estão em desenvolvimento. Isso pode ser um fator não muito interessante para modelos de ML, pois pode causar mais assertividade apenas para essa categoria. Fora este ponto, nota-se também que países em desenvolvimento possuem uma expectativa menor do que países já desenvolvidos (média de 11 anos menor). Ou seja, fatores econômicos também influênciam a expectativa de vida.

```
# Top 10 países com maiores expectativas
categoria = 'Country'
variavel_do_top_10 = 'Life expectancy'

top_10 = life_expectancy_data.groupby(categoria)[variavel_do_top_10].mean().nlargest(10)
top_10 = top_10.sort_values(ascending=True)
top_10.plot(kind='barh', color="#66B2FF")
plt.title(f'Top 10 países com maior expectativa de vida (média acumulada)')
plt.xlabel(f'Média de expectativa de vida')
plt.ylabel(variavel_do_top_10)

plt.show()
```

**Abaixo testei dois modelos de ML:** Um considerando todas as variáveis do dataset, incluindo as categorias Status e Country usando a técnica Dummy. E outro filtrando apenas as variáveis de maior correlação com o alvo (filtrei apenas casos com pelo menos 0.30 / -0.30 de correlação).

### Gerando modelo de ML usando todas as variáveis (Dummy necessário para categorias)

```
# Convertendo variáveis categóricas em dummy para possibilitar usa-las na regressão linear
life_expectancy_dummy = life_expectancy_data
life_expectancy_dummy = pd.get_dummies(life_expectancy_dummy, columns=['Status', 'Country'], drop_first=True)
```

```
# Substituindo os valores nulos pela média, pois o linear_model do sklearn não aceita valores nulos (NaN)
imputer = SimpleImputer(strategy='mean')
life_expectancy_tratado = pd.DataFrame(imputer.fit_transform(life_expectancy_dummy), columns=life_expectancy_dummy.columns)
```

```
# Seperando treino x teste e definindo variável alvo (Life Expectancy)
re_train, re_test = train_test_split(life_expectancy_tratado, train_size=0.7, random_state=42)

X_train = re_train.drop('Life expectancy', axis=1)
y_train = re_train['Life expectancy']

X_test = re_test.drop('Life expectancy', axis=1)
y_test = re_test['Life expectancy']
```

```
X_train.shape
```

```
X_test.shape
```

```
reg = linear_model.LinearRegression()
```

```
reg.fit(X=X_train,y=y_train)
```

```
reg.coef_
```

```
reg.intercept_
```

```
y_pred = reg.predict(X_test)
```

```
mse = metrics.mean_squared_error(y_test, y_pred)

# Calculando rmse
rmse = np.sqrt(mse)

# Média dos valores reais
mean_y = np.mean(y_test)

# RMSE em percentual para facilitar leitura do resultado
rmse_percent = (rmse / mean_y) * 100

print('MSE:', mse)
print('RMSE em %:', rmse_percent)
```

### Gerando modelo de ML com variáveis de maior correlação com alvo

```
# Removendo as colunas categóricas para montar a regressão linear
life_expectancy_filtrado = life_expectancy_data.drop(['Country', 'Status'], axis=1)
```

```
# Substituindo os valores nulos pela média, pois o linear_model do sklearn não aceita valores nulos (NaN)
imputer = SimpleImputer(strategy='mean')
life_expectancy_tratado = pd.DataFrame(imputer.fit_transform(life_expectancy_filtrado), columns=life_expectancy_filtrado.columns)
```

```
life_expectancy_tratado.head(10)
```

```
re_train, re_test = train_test_split(life_expectancy_tratado, train_size=0.7)
X_train = re_train.iloc[:,[2,4,5,8,10,12,13,14,16,17,18,19]]
y_train = re_train.iloc[:,1]

X_test = re_test.iloc[:,[2,4,5,8,10,12,13,14,16,17,18,19]]
y_test = re_test.iloc[:,1]
```

```
X_train.head(10)
```

```
X_test.shape
```

```
reg = linear_model.LinearRegression()
```

```
reg.fit(X=X_train,y=y_train)
```

```
reg.coef_
```

```
reg.intercept_
```

```
y_pred = reg.predict(X_test)
```

```
mse = metrics.mean_squared_error(y_test, y_pred)

# Calculando rmse
rmse = np.sqrt(mse)

# Média dos valores reais
mean_y = np.mean(y_test)

# RMSE em percentual para facilitar leitura do resultado
rmse_percent = (rmse / mean_y) * 100

print('MSE:', mse)
print('RMSE em %:', rmse_percent)
```

### Conclusão

Em termos comparativos, o primeiro modelo apresentou melhores resultados (considerando 100% das variáveis usando Dummy). Para ambos os modelos, é interessante notar que diversos fatores influênciam a expectativa de vida, não somente os dados de saúde. Por exemplo, a escolaridade é a variável de maior relação com o alvo. Este fato fomenta a necessidade procurarmos por dados além da nossa percepção no momento de fazer uma predição. O dataset possui alta correlação entre os dados e não é composto apenas por fatores de saúde (imunização), possui informações inclusive de economia (país desenvolvido x em desenvolvimento) que impacta o resultado final.
